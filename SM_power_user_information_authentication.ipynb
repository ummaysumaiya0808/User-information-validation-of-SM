```python
import cryptography
from cryptography.fernet import Fernet
```

# Function to generate a secret key for encryption

```python
key = Fernet.generate_key()
cipher_text = Fernet(key)
```

```python
plain_text = "Power consumption: 70 kWh"
userID = 'Andrew Boston'
metre_message = f"{plain_text} : {userID}"
print(f"Original message from the smart meter: {metre_message}")
```

# Encrypt the message

```python
encrypted_message = cipher_text.encrypt(metre_message.encode())
print(f"Encrypted message: {encrypted_message}")
```

*Simulate the message being sent to a control center/ utility.*

# Decrypt the received message

```python
decrypted_message = cipher_text.decrypt(encrypted_message).decode()
print(f"Decrypted message at the control center: {decrypted_message}")
```

# Function to authenticate a smart grid device


```python
def authenticate_device(device_id, username, password):
    if device_id in smart_meters:
        # Storing the username and password in the new objects
        stored_username = smart_meters[device_id]["username"]
        stored_password = smart_meters[device_id]["password"]

        if username == stored_username and password == stored_password:
            return True
    return False
```

# Checking the website validation

```python
import socket
```

```python
# Check if a target host is up
def is_host_up(target_host, port_number):
    try:
        # Socket object
        sk = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

        timeout = 2 # Connection attempt (in seconds)

        sk.settimeout(timeout) # Attempt to connect to the target host and port
        sk.connect((target_host, port_number))

        # If the connection is successful, the host is up
        print(f"{target_host}:{port_number} is reachable")

        # Close the socket
        sk.close()

        return True
    except socket.error:
        # If the connection attempt fails, the host is not reachable
        print(f"{target_host}:{port_number} is not reachable")
        return False
```

Your web browser normally establishes a default connection to the server on port 443 when you visit a website via HTTPS. The SSL/TLS encryption protocol, which guarantees that data transferred between your browser and the web server is protected and secure, has allocated this port for secure communication.

# List of trusted IP addresses

The smart grid consists of a home area network (HAN) and wide area network (WAN). Here, we wil, show HAN IP protocol varification with the range of 192.168.0.0 to 192.168.255.255\\\\\

# Generate IP addresses

```python
for a in range(192, 193):  # First octet (192 to 192)
    for b in range(168, 169):  # Second octet (168 to 168)
        for c in range(0, 256):  # Third octet (0 to 255)
            for d in range(0, 256):  # Fourth octet (0 to 255)
                ip_address = f"{a}.{b}.{c}.{d}"
                print(ip_address)
```

```python
HAN_IP = [ip_address]

def is_ip_trusted(ip_address):
    if ip_address in HAN_IP:
        return True
    else:
        return False

if __name__ == "__main__":
    input_ip = input("Enter an IP address to check: ")
    if is_ip_trusted(input_ip):
        print(f"{input_ip} is a trusted IP address. Access granted.")
    else:
        print(f"{input_ip} is not a trusted IP address. Access denied.")
```
